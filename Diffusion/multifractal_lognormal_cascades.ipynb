# From  [Log-Normal continuous cascades: aggregationproperties and estimation.Application to financial time-series](https://arxiv.org/pdf/0804.0185.pdf)


When $\lambda$ is small, MRW is close to its driving Brownian motion, which is consistent since the mean and covariance of $\omega_{l,T}$ tend to zero when $\lambda$ tends to zero, in which case the multifractal measure collapses to the Lebesgue measure, i.e the change of time in the definition of MRW becomes trivial.

As $\lambda$ grows, MRW becomes rougher and seems to become jumpy (which is only a artefact of the numerical discretization, the process actually remains continuous). This effect is even more visible as $\lambda$ reaches $\frac{1}{2}$, which is the limit of convergence for the multifractal measure.

Below is a comparison of MRWs generated with different $\lambda$ and different integral scales. The latter is a decorrelation parameter : the larger it is, the longer the memory in the process, which translates in less frequent large "jumps" and longer period of "flat" signal for values of $\lambda$ close to $\frac{1}{2}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import _LIB_Ito_Diffusion as Ito
plt.style.use('dark_background')
```

```python
T = 1
TAU = 0.1
LEGEND_SIZE = 32
FONT_SIZE = 32
```

```python
fig, axes = plt.subplots(figsize=(21, 42), nrows=5, ncols=1)

intermittencies = [0.02, 0.2, 0.5, np.sqrt(0.5), np.sqrt(0.5) + 0.1]

X = Ito.Lognormal_multifractal(x0=1, T=T, scheme_step=TAU, integral_scale=4)

for i, ax in enumerate(axes):
    intermittency = intermittencies[i]
    X.intermittency = intermittency
    df = X.simulate()
    df.plot(kind='line', ax=ax)

    ax.set_xlabel('time\n$\lambda = ${:.2}\n'.format(intermittency), fontsize=FONT_SIZE)
    ax.set_ylabel('spot', fontsize=FONT_SIZE)
    ax.legend(prop={'size': LEGEND_SIZE})
    ax.tick_params(labelsize=FONT_SIZE)

plt.tight_layout()
plt.show()
```
