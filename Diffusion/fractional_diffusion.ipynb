```python
import numpy as np
import matplotlib.pyplot as plt
from ito_diffusion_1d import *
plt.style.use('dark_background')
```

```python
T = 1
SCHEME_STEPS = int(1e3)
LEGEND_SIZE = 32
FONT_SIZE = 32
```

## Fractional Brownian motions

* $H=\frac{1}{2}$ : Brownian motion
* $H<\frac{1}{2}$ : mean-reverting, rougher sample paths
* $H>\frac{1}{2}$ : trend-following, smoother sample paths


```python
fig, ax = plt.subplots(figsize=(29, 21), nrows=1, ncols=1)

X = FBM(x0=0, T=T, scheme_steps=SCHEME_STEPS, drift=0, vol=1)

Hs = [0.3, 0.4, 0.6, 0.9]

for H in Hs:
    X.H = H
    df = X.simulate()
    df = df.rename(columns={'spot': 'H={:.2f}'.format(H)})    
    df.plot(kind='line', ax=ax)

ax.set_xlabel('time', fontsize=FONT_SIZE)
ax.set_ylabel('spot', fontsize=FONT_SIZE)
ax.legend(prop={'size': LEGEND_SIZE})
ax.tick_params(labelsize=FONT_SIZE)

plt.tight_layout()
plt.show()
```

## Fractional Geometric Brownian motions

```python
fig, ax = plt.subplots(figsize=(29, 21), nrows=1, ncols=1)

X = GBM(x0=1, T=T, scheme_steps=SCHEME_STEPS, drift=0, vol=1)

noise_params = {
    'type': 'fgaussian',
}

Hs = [0.3, 0.4, 0.6, 0.9]

for H in Hs:
    noise_params['H'] = H
    X.noise_params = noise_params
    df = X.simulate()
    df = df.rename(columns={'spot': 'H={:.2f}'.format(H)})    
    df.plot(kind='line', ax=ax)

ax.set_xlabel('time', fontsize=FONT_SIZE)
ax.set_ylabel('spot', fontsize=FONT_SIZE)
ax.legend(prop={'size': LEGEND_SIZE})
ax.tick_params(labelsize=FONT_SIZE)

plt.tight_layout()
plt.show()
```
